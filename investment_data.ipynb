```python
import pandas as pd
import matplotlib.pyplot as plt

def analysis_of_company(s, take_profit):
    
        df = pd.read_csv(s, index_col=0, sep=',') 

        feb = '2023-02-21'

        new_df_idx=[]
    
        for i in df.index:
            if feb in i:
                new_df_idx.append(i)
        
        type(df.index)
        l = df.index.tolist()
        idx = l.index(new_df_idx[0])
        
        cols_name = df.columns
        row_name = df.index.name

        df = pd.read_csv(s, skiprows = idx ,index_col=0, sep=',') 
        df.columns = cols_name
        df.index.name = row_name

        df.index = [x.replace('T14:30:00Z', '') for x in df.index]
        df.index = [x.replace('T13:30:00Z', '') for x in df.index]

        first_date=''
        for i in range (df.shape[0]):
            if df.iloc[i, 1] > take_profit:
                first_date = df.index[i]
                break
        
        title = s.split(',')[0]
        fig, ax = plt.subplots(ncols=1,nrows=1,figsize=[12,8])
        plt.style.use('fivethirtyeight')
        df.loc[:,'high'].plot(kind='line', ax=ax, legend=False, title = title)
        ax.xaxis.set_tick_params(labelrotation=0)
        plt.axhline(y = take_profit, color = 'r', linestyle = '-', linewidth = 2);        
        
```

```python
s = [ 'BATS_NVDA, 1D_0c642.csv', 'BATS_NFLX, 1D_6b1b1.csv', 'BATS_META, 1D_93f25.csv', 'BATS_TSLA, 1D_01f11.csv', 'BATS_GOOG, 1D_b54bc.csv',
     'BATS_MSFT, 1D_7c381.csv', 'BATS_AAPL, 1D_75e8f.csv', 'BATS_DIS, 1D_1de95.csv',  'BATS_COIN, 1D_78e9e.csv']
take_profit = [241.92, 389.53, 193.05, 250, 114, 275.33, 167.43, 120.03, 79.94]

for i in range(len(s)):
# print(s[0], take_profit[0])
    analysis_of_company(s[i],take_profit[i]) 
```
